```python
import pandas as pd
```

```python
def extract_amp_sqamp(input_file, output_file):
    data = []
    
    with open(input_file, 'r', encoding='utf-8') as f:
        for line in f:
            if not line.strip():
                continue
            sections = line.strip().split(':')
            
            amp = sections[-2].strip()
            sqamp = sections[-1].strip()
            
            data.append({'amplitude': amp, 'squared_amplitude': sqamp})
    
    df = pd.DataFrame(data)
    
    df.to_csv(output_file, index=False)
    print(f"Extracted {len(df)} records. Saved to {output_file}")
```

```python
input_file = "QED_data/raw_data.txt" 
output_file = "QED_data/preprocessed_data.csv"
extract_amp_sqamp(input_file, output_file)
```
